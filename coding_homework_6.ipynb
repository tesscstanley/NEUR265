```python
# Step 10: computing the coherence for acc_hits and lc_hits
dt = time[1] - time[0]              # Define the sampling interval.
#K = acc_hits.shape[0]               # Define the number of trials.
N = lc_hits.shape[1]               # Define number of points in each trial.
#ac = np.zeros([2 * N - 1])       # Declare empty vector for autocov.
  

# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials

Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum


# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real, color = 'red')                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between ACC & LC')
```

#### Step 11: describing results, in which frequency band is there the most phase coherence during hits?

The frequency band where there is the most phase coherence is at around 1Hz. Another tall peak of phase coherence is around 6Hz as well. However, as frequency increase the amount of phase coherence decreases between the acc_hits and lc_hits data. 

```python
# Step 12: repeating step 10, but with acc_fa and lc_fa in blue plotted with acc_hits and lc_hits in red

dt = time[1] - time[0]              # Define the sampling interval.
K = acc_fa.shape[0]               # Define the number of trials.
N = lc_fa.shape[1]               # Define number of points in each trial.
ac = np.zeros([2 * N - 1])       # Declare empty vector for autocov.


# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T

# for False Alarms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  # ... and each trial in E2
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# for Hits
xf2 = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf2 = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2
Hxx = scale * (xf2 * xf2.conj()).mean(0)  # Spectrum of E1 trials
Hyy = scale * (yf2 * yf2.conj()).mean(0)  # ... and E2 trials
Hxy = scale * (xf2 * yf2.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))
cohr2 = np.abs(Hxy) / (np.sqrt(Hxx) * np.sqrt(Hyy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.

plt.plot(f, cohr.real, color = 'blue')      # Plot coherence vs frequency,
plt.plot(f, cohr2.real, color = 'red')
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between ACC & LC')
```
