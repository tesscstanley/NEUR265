```python
# Purpose: to compare aSyn area analyzed and average GFAP positive cells across CERAD scores
# DV1 = aSyn area analyzed      DV2 = average GFAP positive cell area      IV = CERAD scores
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# data1 = aSyn and GFAP data with CERAD scores
data1 = pd.read_csv('https://raw.githubusercontent.com/tesscstanley/NEUR265/main/Version%20B%3A%20aSyn%20%2B%20GFAP%20%2B%20CERAD.csv')

```

```python
# Sorting data across the CERAD scores: absent, sparse, moderate, frequent
absent = data1.loc[data1['CERAD score']=='Absent']
sparse = data1.loc[data1['CERAD score']=='Sparse']
moderate = data1.loc[data1['CERAD score']=='Moderate']
frequent = data1.loc[data1['CERAD score']=='Frequent']
```

#Data/Graphs for WHOLE data

```python
# Creating a bar graph to visualize the distribution of data points across CERAD scores as groups
#2/27
data1['CERAD score'].value_counts().loc[['Absent','Sparse','Moderate','Frequent']].plot(kind='bar', color='darkseagreen')
plt.ylabel('Frequency')
plt.title('Distribution of CERAD Scores')
```

```python
# Creating a scatter plot to show the relationship between the aSyn area analyzed and average GFAP positve cell area, with dots color-coded by CERAD score
#2/15
plt.scatter(absent['aSyn area analyzed'],absent['average GFAP positive branch area'], color = 'cornflowerblue')
plt.scatter(sparse['aSyn area analyzed'], sparse['average GFAP positive branch area'], color = 'plum')
plt.scatter(moderate['aSyn area analyzed'], moderate['average GFAP positive branch area'], color = 'darkseagreen')
plt.scatter(frequent['aSyn area analyzed'], frequent['average GFAP positive branch area'], color = 'lightcoral')

plt.ylabel('Average GFAP Positive Branch Area')
plt.xlabel('aSyn Positive Branch Area')
plt.legend(['Absent','Sparse','Moderate','Frequent'])
plt.title('Relationship Between Avg. GFAP Positive Area and aSyn Positive Area by CERAD Score')
```

#Data/Graphs For aSyn

```python
# Creating a boxplot visualizing how aSyn area analyzed distribution differs between CERAD scores

absent_data1 = data1[data1['CERAD score'] == 'Absent']
absent_bp = absent_data1['average aSyn positive cell area']

sparse_data = data1[data1['CERAD score'] == 'Sparse']
sparse_bp = sparse_data['average aSyn positive cell area']


moderate_data = data1[data1['CERAD score'] == 'Moderate']
moderate_bp = moderate_data['average aSyn positive cell area']

frequent_data = data1[data1['CERAD score'] =='Frequent']
frequent_bp = frequent_data['average aSyn positive cell area']
#2/27

aSynboxdata = [absent_bp.tolist(),sparse_bp.tolist(),moderate_bp.tolist(),frequent_bp.tolist()]
plt.boxplot(aSynboxdata)
plt.xticks([1,2,3,4],['Absent','Sparse','Moderate','Frequent'])
plt.xlabel('CERAD score')
plt.ylabel('Average GFAP')
plt.title('Average aSYN Distribution Across CERAD Scores')
```

```python
# Creating a boxplot visualizing how aSyn area analyzed distribution differs between CERAD scores

absent_data = data1[data1['CERAD score'] == 'Absent']
absent_bp = absent_data['aSyn area analyzed']

sparse_data = data1[data1['CERAD score'] == 'Sparse']
sparse_bp = sparse_data['aSyn area analyzed']


moderate_data = data1[data1['CERAD score'] == 'Moderate']
moderate_bp = moderate_data['aSyn area analyzed']

frequent_data = data1[data1['CERAD score'] =='Frequent']
frequent_bp = frequent_data['aSyn area analyzed']

#2/27
aSynboxdata = [absent_bp.tolist(),sparse_bp.tolist(),moderate_bp.tolist(),frequent_bp.tolist()]
plt.boxplot(aSynboxdata)
plt.xticks([1,2,3,4],['Absent','Sparse','Moderate','Frequent'])
plt.xlabel('CERAD score')
plt.ylabel('Average aSyn')
plt.title('Average aSyn Area Distribution Across CERAD Scores')
```

```python
# Visualizing the distribution of aSyn area analyzed in each CERAD Score
fig, ax = plt.subplots(2,2,figsize=(20,7))

CERADscores = data1['CERAD score'].unique()
for Cscore in CERADscores:

    cs = data1[data1['CERAD score'] == Cscore]
    
    if Cscore == 'Absent':
      ax[0,0].hist(cs['average aSyn positive cell area'],
                label=Cscore,color="cornflowerblue")
      ax[0,0].set_title('Absent')
      ax[0,0].set_ylabel('Frequency')
      
    elif Cscore == 'Sparse':
      ax[1,0].hist(cs['average aSyn positive cell area'],
                label=Cscore,color="plum")
      ax[1,0].set_title('Sparse')
      ax[1,0].set_ylabel('Frequency')
      ax[1,0].set_xlabel('aSyn Positive Branch Area')
      
    elif Cscore == 'Moderate':
      ax[0,1].hist(cs['average aSyn positive cell area'],
                label=Cscore,color="darkseagreen")
      ax[0,1].set_title('Moderate')
      ax[0,1].set_ylabel('Frequency')
      
      
    elif Cscore == 'Frequent':
      ax[1,1].hist(cs['average aSyn positive cell area'],
                label=Cscore,color="lightcoral")
      ax[1,1].set_title('Frequent')
      ax[1,1].set_ylabel('Frequency')
      ax[1,1].set_xlabel('aSyn Positive Branch Area')
```

```python
# data2 = AT8 and RIPA pTau data in Braak stages
data2 = pd.read_csv('https://raw.githubusercontent.com/tesscstanley/NEUR265/main/AT8%20%2B%20pTau%20%2B%20Braak.csv')
```

```python
# Creating a bar graph to visualize the distribution of data points across Braak stages as groups
#2/27
data2['Braak stages'].value_counts().loc[['Braak 0','Braak II','Braak III','Braak IV','Braak V','Braak VI']].plot(kind='bar', color='darkseagreen')
# Braak I was not included as no sample was in the Braak I stage
plt.ylabel('Frequency')
plt.title('Distribution of Braak Stages')
```

## AT8 data

```python
# Creating a boxplot visualizing how average AT8 positive cell area distribution differs between Braak staging

B0_data = data2[data2['Braak stages'] == 'Braak 0']
B0_bp = B0_data['average AT8 positive cell area']

BII_data = data2[data2['Braak stages'] == 'Braak II']
BII_bp = BII_data['average AT8 positive cell area']

BIII_data = data2[data2['Braak stages'] == 'Braak III']
BIII_bp = BIII_data['average AT8 positive cell area']

BIV_data = data2[data2['Braak stages'] == 'Braak IV']
BIV_bp = BIV_data['average AT8 positive cell area']

BV_data = data2[data2['Braak stages'] == 'Braak V']
BV_bp = BV_data['average AT8 positive cell area']

BVI_data = data2[data2['Braak stages'] == 'Braak VI']
BVI_bp = BVI_data['average AT8 positive cell area']

#2/27
aSynboxdata = [B0_bp.tolist(),BII_bp.tolist(),BIII_bp.tolist(),BIV_bp.tolist(),BV_bp.tolist(),BVI_bp.tolist()]
plt.boxplot(aSynboxdata)
plt.xticks([1,2,3,4,5,6],['Braak 0','Braak II','Braak III','Braak IV','Braak V','Braak VI'])
plt.xlabel('Braak Staging')
plt.ylabel('Average AT8 Positive Cell Area')
plt.title('Average AT8 Positive Cell Area Distribution Across Braak Staging')
```
