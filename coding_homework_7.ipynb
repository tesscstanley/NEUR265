```python
# Step 2: Importing numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 3: importing spikes_left_trial1.csv, spikes_right_trial1.csv, spikes_left_trial2, and spikes_right_trial2 as numpy arrays 
# called spikes_left_trial1, spikes_right_trial1, spikes_left_trial2, and spikes_right_trial2, respectively
url1 = 'https://raw.githubusercontent.com/tesscstanley/NEUR265/main/spikes_left_trial1.csv'
spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')

url2 = 'https://raw.githubusercontent.com/tesscstanley/NEUR265/main/spikes_right_trial1.csv'
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')

url3 = 'https://raw.githubusercontent.com/tesscstanley/NEUR265/main/spikes_left_trial2.csv'
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')

url4 = 'https://raw.githubusercontent.com/tesscstanley/NEUR265/main/spikes_right_trial2.csv'
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
# Step 4: printing the length of each array
print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
# Step 6: determining the firing rate of each trial
#left trail 1
print((len(spikes_left_trial1))/30)
#left trail 2
print((len(spikes_left_trial2))/30)
#right trail 1
print((len(spikes_right_trial1))/30)
#right trail 2
print((len(spikes_right_trial2))/30)
```

#### Step 7: interpreting firing rate across left and right trails
Since my data across left and right trails and across trials 1 and 2 for both turns is very variable, I do not see a pattern. #### Step 9: interpreting the graph from step 8, does this neuron seem more "bursty", or more regular-firing, during left or right turn trials?
Between the left and right turn trials for both trial 1 and 2, the neuron appears more "bursty" during left turns, as there are a greater number of periods with rapid firing close together, followed by periods of no activity. For right turns, there appears to be more spaced out and regular firing, although towards the end of trial 1 for left turns, there appears to be a small period of regular firing. However, the beginning of the waiting period before a right turn in trial 2 appears to display some bursty firing. This may align with how the firing rates for right turns were higher in the second trial than the first trial. 


```python
# Step 10: calculating inter-spike interval (ISI) arrays for each spike arrays using the np.diff() function
left1 = np.diff(spikes_left_trial1)
print(left1)
right1 = np.diff(spikes_right_trial1)
print(right1)
left2 = np.diff(spikes_left_trial2)
print(left2)
right2 = np.diff(spikes_right_trial2)
print(right2)
```

```python
# Step 11: making four histograms for the four ISI arrays, with bins starting at time 0, and ending at time 2 s, with a bin size equal to 0.01 s
binsize = 0.01
trialbins11 = np.arange(0,2,binsize)
fig,ax = plt.subplots(2,2,figsize=(20,7))
ax[0,0].hist(left1, bins=trialbins11)
ax[0,0].set_xlim([-0.1, 2])
ax[0,0].set_ylim([0, 8])
ax[0,0].set_xlabel('Time [s]')
ax[0,0].set_ylabel('Firing')
ax[0,0].set_title('Left')

ax[0,1].hist(right1, bins=trialbins11)
ax[0,1].set_xlim([-0.1, 2])
ax[0,1].set_ylim([0, 8])
ax[0,1].set_xlabel('Time [s]')
ax[0,1].set_ylabel('Firing')
ax[0,1].set_title('Right')

ax[1,0].hist(left2, bins=trialbins11)
ax[1,0].set_xlim([-0.1, 2])
ax[1,0].set_ylim([0, 8])
ax[1,0].set_xlabel('Time [s]')
ax[1,0].set_ylabel('Firing')

ax[1,1].hist(right2, bins=trialbins11)
ax[1,1].set_xlim([-0.1, 2])
ax[1,1].set_ylim([0, 8])
ax[1,1].set_xlabel('Time [s]')
ax[1,1].set_ylabel('Firing')
  
```

#### Step 12: interpreting histograms, does this neuron tend to fire in "bursts"? Does it have higher "burstiness" during left or right turn trials?

Yes, this neuron tends to fire in bursts, as one can see how this neuron tends to fire rapidly and close together, followed by bouts of no firing or more sporatic firing. In tiral 1, the neuron appears to be more bursty during left turns compared to right turns, however during trial 2, the neuron appears to be more bursty during the right turns. This is consistent with my data from the firing rate calculation and spike trains. 


```python
# Step 13: finding firing rate across time and then creating 2 subplot histograms with bins starting with 0, ending at 30, size = 1
binsize = 1
trialbins13 = np.arange(0,30,binsize)
# Left Trial 1
# np.zeros makes an array of defined size, populating it with zeros, initializing the variable, empty variable before you start putting data in it


FRleft1 = np.zeros(np.size(trialbins13)-1) 
for i in range(len(trialbins13)-1):        
  spk_left1 = (spikes_left_trial1 > trialbins13[i]) & (spikes_left_trial1 < trialbins13[i+1])
  true_values = spikes_left_trial1[spk_left1]
  FRleft1[i] = (len(true_values))/binsize 

# Right Trial 1
FRright1 = np.zeros(np.size(trialbins13)-1) 
for i in range(len(trialbins13)-1):        
  spk_right1 = (spikes_right_trial1 > trialbins13[i]) & (spikes_right_trial1 < trialbins13[i+1])
  true_values = spikes_right_trial1[spk_right1]
  FRright1[i] = (len(true_values))/binsize

# Left Trial 2
FRleft2 = np.zeros(np.size(trialbins13)-1) 
for i in range(len(trialbins13)-1):        
  spk_left2 = (spikes_left_trial2 > trialbins13[i]) & (spikes_left_trial2 < trialbins13[i+1])
  true_values = spikes_left_trial2[spk_left2]
  FRleft2[i] = (len(true_values))/binsize

# Right Trial 2
FRright2 = np.zeros(np.size(trialbins13)-1)
for i in range(len(trialbins13)-1):         
  spk_right2 = (spikes_right_trial2 > trialbins13[i]) & (spikes_right_trial2 < trialbins13[i+1])
  true_values = spikes_right_trial2[spk_right2]
  FRright2[i] = (len(true_values))/binsize

# subplots
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(trialbins13[:-1], FRleft1, color = 'red')
ax[0].plot(trialbins13[:-1], FRright1, color = 'blue')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')

ax[1].plot(trialbins13[:-1], FRleft2, color = 'red')
ax[1].plot(trialbins13[:-1], FRright2, color = 'blue')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')

```

Yes, the firing rate differs between the right and left turns. In trial 1, right trials appear to have higher firing rate towards the beginning of the 30-second waiting period, while the left turns shows higher firing rates around 10-17 seconds, where right turn firing is low.
