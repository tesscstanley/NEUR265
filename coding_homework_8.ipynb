```python
# Step 12: creating the predictors_conditioning variable that contains shock onset and tone onset
# additional imports from 04/12 for the code below:
from statsmodels.genmod.families import Poisson
import statsmodels.api as sm
from pandas import DataFrame as df

# creating the first trace for the conditioning session
first_trace = conditioning_traces[:,0]
# Re-creating shock_array and tone_array variables for 10458 samples (# of samples in conditioning_traces)
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0


# creating the predictors_conditioning variable with the shock onset and tone onset as predictors for the conditioning session
predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
modelC = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
modelC_results = modelC.fit() # Fit model to our data
```

```python
# Step 13: creating  r_model_conditioning variable using np.zeros, with changed # of ROIs for np.zeros and range()
first_trace = conditioning_traces[:,0]
predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

# creating *for loop* to loop through all of the ROIs in conditioning_traces array, run sm.GLM between the trace and the predictors variable, fit the model,
# create a predicted trace, get a correlation coefficient using pearsonr, and store the correlation coefficient r_index_conditioning columns
for i in range(63):
  trace_temp = conditioning_traces[:,i]
  modelC = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  modelC_results = modelC.fit()
  predicted_trace = modelC_results.predict()
  rC = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = rC.statistic
```

```python
# Step 14: repeating steps 12 and 13 for extinction_traces, creating variables r_index_extinction and extinction_predictors 
# creating the first trace for extinction session
first_trace = extinction_traces[:,0]
# Re-creating shock_array and tone_array variables for 14209 samples (# of samples in extinction_traces)
shock_array = np.zeros(14209)
for i in range(14209):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0


# creating predictors_extinction variable with the tone onset as predictor for the extinction session
predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
modelE = sm.GLM(first_trace, predictors_extinction, family=Poisson())
modelE_results = modelE.fit() # Fit model to our data
# creating the r_index_extinction variable using np.zeros, with changed # of ROIs for np.zeros and range()
r_index_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  modelE = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  modelE_results = modelE.fit()
  predicted_trace = modelE_results.predict()
  rE = pearsonr(trace_temp, predicted_trace)
  r_index_extinction[i] = rE.statistic
```

```python
# Step 15: using the pearsonr function to correlate r_index_conditioning and r_index_extinction variables
pearsonr(r_model_conditioning,r_index_extinction) 
```

```python
# Step 16: creating a scatter plot with r_index_conditioning on the x-axis and r_index_extinction on the y-axis
plt.scatter(r_model_conditioning, r_index_extinction)
plt.xlabel('Conditioning Traces')
plt.ylabel('Extinction Traces')
```
